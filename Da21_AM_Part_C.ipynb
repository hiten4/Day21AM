### Q1 — What is NumPy broadcasting? Why is it useful?

NumPy broadcasting is a mechanism that allows operations between arrays of different shapes without explicitly reshaping them.

It works by automatically expanding the smaller array to match the larger array during operations.

Why useful:

* Eliminates need for loops
* Makes code concise and readable
* Improves performance due to vectorization

Example:
Adding a 1D array to each row of a 2D array.

### Q2 — normalize(X) function

```python
import numpy as np

def normalize(X):
    min_val = np.min(X)
    max_val = np.max(X)
    return (X - min_val) / (max_val - min_val)


# Example
X = np.array([10, 20, 30, 40])
print(normalize(X))
```



### Q3 — Difference between vectorisation and loops

Vectorisation means performing operations on entire arrays at once using NumPy functions, without explicit loops.

Loops process elements one by one using Python iteration.

Why NumPy is faster:

* Uses optimized C-level implementation
* Avoids Python loop overhead
* Performs batch operations efficiently
