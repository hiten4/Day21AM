# 1. Matrix operations

```python
import numpy as np
import time



A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

matmul = np.dot(A, B)


transpose_A = A.T


det_A = np.linalg.det(A)

print("Matrix Multiplication:\n", matmul)
print("Transpose:\n", transpose_A)
print("Determinant:", det_A)
```

# 2. Solve linear system Ax = b

```python
A = np.array([[2, 3],
              [4, 1]])

b = np.array([8, 10])

solution = np.linalg.solve(A, b)

print("Solution:", solution)
```

# 3. Performance comparison

```python

large_array = np.random.rand(1000000)


start = time.time()
total = 0
for x in large_array:
    total += x
loop_time = time.time() - start


start = time.time()
np_total = np.sum(large_array)
numpy_time = time.time() - start

print("Loop time:", loop_time)
print("NumPy time:", numpy_time)
print("Difference:", loop_time - numpy_time)
```
